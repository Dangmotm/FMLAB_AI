```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import math

from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score

diabetes = datasets.load_diabetes()
print("Số chiều dữ liệu input: ", diabetes.data.shape)
print("Kiểu dữ liệu input: ", type(diabetes.data))
print("Số chiều dữ liệu target: ", diabetes.target.shape)
print("Kiểu dữ liệu target: ", type(diabetes.target))
print()
print("5 mẫu dữ liệu đầu tiên:")
print("input: ", diabetes.data[:5])
print("target: ",diabetes.target[:5])
```

```python
X = diabetes.data
y = diabetes.target

X_train = X[:361]
y_train = y[:361]

X_test = X[362:]
y_test = y[362:]

from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression, Ridge

sc = StandardScaler()
sc.fit(X_train)
X_train = sc.transform(X_train)
X_test = sc.transform(X_test)

lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
print("[w1, w2, ..., w_n] = ", lr.coef_)
print("w0 = ", lr.intercept_)
```

```python
lr_score = lr.score(X_test, y_test)
lr_score
```

```python
y_pred = lr.predict(X_test)
pd.DataFrame(data = np.array([y_test, y_pred, abs(y_test - y_pred)]).T, columns = ["Thực tế", "Dự đoán", "Lệch"])

y_pred1 = y_pred
```

```python
ridge = Ridge(alpha = 0.1)
ridge.fit(X_train, y_train)
```

```python
print("[w1, w2, ..., w_n] = ", ridge.coef_)
print("w0 = ", ridge.intercept_)
```

```python
ridge_score = ridge.score(X_test, y_test)
ridge_score
```

```python
print("Giá trị thực tế: ", y_test[0])

y_pred_lr = lr.predict(X_test[0:1])
print("Giá trị dự đoán cho mô hình linear regression: ", y_pred_lr)

y_pred_lr_0 = sum(lr.coef_ * X_test[0]) + lr.intercept_
print("Giá trị dự đoán cho mô hình linear regression theo công thức: ", y_pred_lr_0)

y_pred_ridge = ridge.predict(X_test[0:1])
print("Giá trị dư đoán cho mô hình rigde regression: ", y_pred_ridge)

y_pred_ridge_0 = sum(ridge.coef_ * X_test[0]) + ridge.intercept_
print("Giá trự dự đoán cho mô hình ridge rgression theo công thức: ", y_pred_ridge_0)
```

```python
y_pred = ridge.predict(X_test)

pd.DataFrame(np.array([y_test, y_pred, abs(y_test - y_pred)]).T, columns = ["Thực tế", "Dự đoán", "Lệch"])
```

```python
math.sqrt(mean_squared_error(y_test, y_pred))
```

```python
_lambda = [0, 0.0001,0.01, 0.04, 0.05, 0.06, 0.1, 0.5, 1, 5, 10, 20]
RMSE = []

for i in _lambda:
    ridge = Ridge(alpha = i)
    ridge.fit(X_train, y_train)

    y_pred = ridge.predict(X_test)
    ridge_rmse = math.sqrt(mean_squared_error(y_test, y_pred))
    RMSE.append(ridge_rmse)

    print("Lambda: ", i, "; RMSE = ", ridge_rmse)
```
