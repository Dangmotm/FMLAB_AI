```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
```

```python
n_samples = 100
random_state = 170
center_points = [[1, 1], [-1, -1], [1, -1]]

X, y = make_blobs(n_samples = n_samples, random_state = random_state, centers = center_points, cluster_std = 0.6)

print("Số chiều dữ liệu: ", X.shape, y.shape)
print("5 điểm dữ liệu đầu tiên :\n", X[:5])
```

```python
plt.figure(figsize = (12, 12))
plt.subplot(221)
plt.scatter(X[:, 0], X[:, -1], c = 'blue')
plt.title("Các điểm dữ liệu trước khi phân cụm, Số lượng: {}".format(n_samples))
plt.show()
```

```python
k_cluster = 3
k_mean = KMeans(n_clusters = k_cluster, random_state = random_state)
k_mean.fit(X)

centers = np.array(k_mean.cluster_centers_)

print("Tâm cụm sau khi training ({} tâm): \n".format(k_cluster), centers)
```

```python
y_pred = k_mean.predict(X)
print("Kết quả dự đoán cho 5 mẫu dữ liệu đầu tiên trong tập data: \n")
print(y_pred[:5])
```

```python
plt.figure(figsize = (12, 12))
plt.subplot(221)
plt.scatter(X[:, 0], X[:, 1], c = y_pred)
plt.scatter(centers[:, 0], centers[:, 1], c = "red")
plt.title("Các điểm dữ liệu sau khi phân cụm")
plt.show()
```

```python
n_samples = 100
random_state = 170
center_points = [[-1.8, -1.8], [3.6, 3.6]]

X, y = make_blobs(n_samples = n_samples, random_state = random_state, centers = center_points, cluster_std = 0.6)

print("Số chiều dữ liệu: ", X.shape, y.shape)
print("5 điểm dữ liệu đầu tiên :\n", X[:5])
```

```python
plt.figure(figsize = (12, 12))
plt.subplot(221)
plt.scatter(X[:, 0], X[:, -1], c = 'blue')
plt.title("Các điểm dữ liệu trước khi phân cụm, Số lượng: {}".format(n_samples))
plt.show()
```

```python
k_cluster = [3, 4, 5]
plt.figure(figsize = (20, 3))
for i in k_cluster:
    k_mean = KMeans(n_clusters = i, random_state = random_state)
    k_mean.fit(X)

    centers = np.array(k_mean.cluster_centers_)
    y_pred = k_mean.predict(X)

    plt.subplot(1, 3, i - 2)
    plt.scatter(X[:, 0], X[:, 1], c = y_pred)
    plt.scatter(centers[:, 0], centers[:, 1], c = "red")
    plt.title("Các điểm dữ liệu sau khi phân thành {} cụm".format(i))

plt.show()
```
