```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("cancer_data.csv")
df.head()
```

```python
df.info()
```

```python
df = df.drop(["id", "Unnamed: 32"], axis = 1)

M = df[df.diagnosis == 'M']
B = df[df.diagnosis == 'B']

M
```

```python
B
```

```python
plt.figure(figsize = (16, 9))
plt.title("Trực quan dữ liệu u ác tính và u lành")
plt.xlabel("Radius mean")
plt.ylabel("Texture mean")
plt.scatter(M.radius_mean, M.texture_mean, color = "red", alpha = 0.5, label = "U ác")
plt.scatter(B.radius_mean, B.texture_mean, color = "cyan", alpha = 0.5, label = "U lành")
plt.legend()
plt.show()
```

```python
df.diagnosis = [1 if i == 'M' else 0 for i in df.diagnosis]
df
```

```python
X = df.drop(["diagnosis"], axis = 1)
y = df.diagnosis.values
X
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
sc.fit(X)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 18)
```

```python
from sklearn.naive_bayes import MultinomialNB

model_MNB = MultinomialNB()
model_MNB.fit(X_train, y_train)

print("Multinormial Naive Bayes score: ", model_MNB.score(X_test, y_test))
```

```python
from sklearn.naive_bayes import GaussianNB

model_GNB = GaussianNB(var_smoothing = 1e-3)
model_GNB.fit(X_train, y_train)

print("Gaussian Naive Bayes score: ", model_GNB.score(X_test, y_test))
```

```python
X_test = X_test.reset_index(drop = True)

test = X_test.iloc[0:1]
test
```

```python
print('Label: ', y_test[0])
print("________________")
print('Predict: ', model_GNB.predict(test)[0])
```
