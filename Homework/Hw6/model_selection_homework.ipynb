```python
from tqdm import tqdm
from sklearn.model_selection import learning_curve, train_test_split, GridSearchCV
from sklearn.pipeline import Pipeline

from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.ensemble import RandomForestRegressor
%matplotlib inline

from sklearn.metrics import get_scorer_names
get_scorer_names()

seed = 42
```

```python
X = df.drop(['id', 'date', 'price'], axis = 1)
y = df['price']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = seed)

print('Dữ liệu Training: ', X_train.shape, y_train.shape)
print('Dữ liệu Testing: ', X_test.shape, y_test.shape)
```

```python
def cross_validation(estimator):
    _, train_scores, test_scores = learning_curve(estimator, X_train, y_train, cv = 10, 
                                                  n_jobs = -1, train_sizes = [1.0, ], scoring = 'neg_mean_absolute_error')
    test_scores = test_scores[0]
    mean = test_scores.mean()
    std = test_scores.std()
    return mean, std
```

```python
def plot(title, xlabel, X, Y, error, ylabel = 'mean_squared_error'):
    plt.xlabel(xlabel)
    plt.title(title)
    plt.ylabel(ylabel)
    plt.grid()
    plt.errorbar(X, Y, error, linestyle = 'None', marker = 'o')
```

```python
title = 'Thay đổi alpha'
xlable = 'Alpha'

X, Y = [], []
error = []

for a in tqdm([0, 0.0001, 0.001, 0.01, 0.1, 1, 10, 100]):
    lasso = Lasso(alpha = a)
    mean, std = cross_validation(lasso)
    X.append(str(a))
    Y.append(mean)
    error.append(std)

plot(title, xlable, X, Y, error)
plt.show()
```

```python
title = 'Thay đổi số lượng láng giềng'
xlable = 'Số lượng láng giềng'

X, Y = [], []
error = []

for k in tqdm([1, 2, 3, 5, 7, 9, 11, 15, 20, 25, 50]):
    knn = KNeighborsRegressor(n_neighbors = k)
    mean, std = cross_validation(knn)
    X.append(str(k))
    Y.append(mean)
    error.append(std)

plot(title, xlable, X, Y, error)
plt.show()
```

```python
def grid_search(model, dict_param):    
    classifier = GridSearchCV(estimator = model, param_grid = dict_param, n_jobs = -1, verbose = 1)
    classifier.fit(X_train, y_train)
    
    print(classifier.best_estimator_)
    return classifier.best_estimator_
```
